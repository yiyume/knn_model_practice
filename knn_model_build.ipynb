## k - nearest neighbors
this algorithm has a very serious weakness, even use the train data if the data lies in the boundary, the prediction will get different result.

```python
import numpy as np
import pandas as pd
from collections import Counter

df = pd.read_csv('./KNNAlgorithmDataset.csv')
```

```python
# select some columns as X and y
train_data_all = df[['diagnosis','radius_mean','texture_mean','perimeter_mean','area_mean']]
```

```python
train_data_all.head()
```

```python
train_data_xy = train_data_all.iloc[:565]
```

```python
features = list(train_data_all.columns)
features.remove('diagnosis')

train_data = train_data_xy[features].to_numpy()
train_labels = train_data_xy['diagnosis'].to_numpy()
```

```python
# define a Euclidean distance function
def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((np.array(point1) - np.array(point2))**2))

def knn_predict(train_data, train_labels, test_point, k):
    distances = []
    for i in range(len(train_data)):      
        dist = euclidean_distance(train_data[i], test_point)
        distances.append((dist, train_labels[i]))
    distances.sort(key=lambda x : x[0])
    k_nearest_labels = [label for _, label in distances[:k]]
    return Counter(k_nearest_labels).most_common(1)[0][0]
```

```python
prediction = knn_predict(train_data,train_labels,train_data_all.iloc[560,1:],5)
print(prediction)
```
